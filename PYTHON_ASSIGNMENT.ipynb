Given a two list. Create a third list by picking an odd-index element from the first list and even index elements from the second.

```python
a = [6,7,8,9,2,4,5]
b = [45,78,65,34,2,3]
even_list = []
odd_list = []

for i in range(len(a)):
  if i % 2 == 1:
    odd_list.append(a[i])
for i in range(len(b)):
  if i % 2 == 0:
    even_list.append(b[i])
print(" even index value : ", even_list)
print("odd index value :", odd_list)
c =[]
for i in odd_list:
  c.append(i)
for j in even_list:
  c.append(j)
print("Final list : ",c)
```

```python
list1=[1,2,3]
print("original list: ",list1)
list2=[]
for i in list1:
    for j in list1:
         for k in list1:
                list2.append([i] + [j] + [k])
print("\n all possible permutation: ",list2)
```
