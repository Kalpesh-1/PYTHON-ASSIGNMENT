# Q.5) Write a Python program to generate all permutations of a list in Python.

```python
list1 = [24,45,23,5,53]
print('list: ',list1)
list2=[]
for i in list1:
    for j in list1:
         for k in list1:
                list2.append([i]+[j]+[k])
print(list2)
```

# Q.10) Write a Python program to remove duplicates from a list of lists. 
#Sample list : [[10, 20], [40], [30, 56, 25], [10, 20], [33], [40]]

```python
l=[]
a = [[10,20],[40],[30,56,25],[10,20],[33],[40]]
for num in a:
  if num not in l:
    l.append(num)
print(l)
```
